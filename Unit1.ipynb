```python
#Strings
s = "Hello, World!"
print(s[0])              # H  (indexing)
print(s[0:5])           # Hello (slicing)
print(s.upper())         # HELLO, WORLD!
print(s.lower())         # hello, world!
print(s.replace("World","Python")) # Hello, Python!
print(s.split(","))      # ['Hello', ' World!']
print(s.strip())         # removes leading/trailing spaces
print(len(s))            # 13
print(f"Length: {len(s)}")  # f-string formatting
print("Ha" * 3)          # HaHaHa
```

```python
#Arithmetic Operator
x, y = 10, 3
print(x + y)    # 13  — addition
print(x - y)    # 7   — subtraction
print(x * y)    # 30  — multiplication
print(x / y)    # 3.333 — division (float)
print(x // y)   # 3   — floor division
print(x % y)    # 1   — modulus (remainder)
print(x ** y)   # 1000 — exponentiation
```

```python
#Comparison Operator
x = 5; y = 8
x == y   # False  equal
x != y   # True   not equal
x < y    # True   less than
x > y    # False  greater
x <= y   # True   less/equal
x >= y   # False  greater/equal
```

```python
#Identify & Membership Operator
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)        # True  — same object in memory
print(a is c)        # False — different objects
print(a is not c)   # True

print(2 in a)        # True  — membership
print(5 not in a)   # True
```

```python
#Decision Making (if / elif / else) branching
marks = 75

if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")   # ← this runs
elif marks >= 60:
    print("Grade: C")
else:
    print("Grade: F")

# One-liner (ternary)
result = "Pass" if marks >= 40 else "Fail"
```

```python
#for Loop
# Loop over range
for i in range(1, 6):
    print(i, end=" ")
# 1 2 3 4 5

# Loop over list
fruits = ["apple","mango","banana"]
for f in fruits:
    print(f)

# enumerate — index + value
for i, f in enumerate(fruits):
    print(i, f)
```

```python
#while Loop
count = 1
while count <= 5:
    print(count)
    count += 1

# while with else
n = 3
while n > 0:
    print(n)
    n -= 1
else:
    print("Done!")
```

```python
#Loop Control Statements break / continue / pass
# break — exit loop immediately
for i in range(10):
    if i == 5:
        break
    print(i, end=" ")   # 0 1 2 3 4

# continue — skip current iteration
for i in range(6):
    if i % 2 == 0:
        continue
    print(i, end=" ")   # 1 3 5  (skips even)

# pass — placeholder (does nothing)
for i in range(3):
    pass   # empty loop — no error
```

```python
#User-Defined Functions def
def greet(name):
    """Returns a greeting message."""
    return f"Hello, {name}!"

print(greet("Alice"))   # Hello, Alice!

# Function with multiple return values
def min_max(lst):
    return min(lst), max(lst)

lo, hi = min_max([3, 1, 7, 2])
print(lo, hi)   # 1 7
```

```python
#Function Argument Types(arguments)
# 1. Positional arguments
def add(a, b):
    return a + b
add(3, 5)   # 8

# 2. Default arguments
def power(base, exp=2):
    return base ** exp
power(4)      # 16 (uses default exp=2)
power(2, 10)  # 1024

# 3. Keyword arguments
def info(name, age):
    print(f"{name} is {age}")
info(age=20, name="Bob")  # order doesn't matter

# 4. *args — variable positional
def total(*nums):
    return sum(nums)
total(1, 2, 3, 4)   # 10

# 5. **kwargs — variable keyword
def profile(**data):
    for k, v in data.items():
        print(f"{k}: {v}")
profile(name="Alice", city="Raipur", age=21)
```

```python
#Lambda (Anonymous) Functions lambda
# lambda arguments: expression
square = lambda x: x ** 2
print(square(5))   # 25

# Common use — sorting
students = [("Alice",85), ("Bob",72), ("Carol",91)]
students.sort(key=lambda s: s[1], reverse=True)
print(students)
# [('Carol', 91), ('Alice', 85), ('Bob', 72)]

# with map() and filter()
nums = [1,2,3,4,5]
squared  = list(map(lambda x: x**2, nums))   # [1,4,9,16,25]
evens    = list(filter(lambda x: x%2==0, nums))# [2,4]
```
