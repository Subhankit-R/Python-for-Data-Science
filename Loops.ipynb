```python
#if / elif / else — decision making
# Basic if
age = 18
if age >= 18:
    print("Adult")         # runs — condition is True

# if-else
num = 7
if num % 2 == 0:
    print("Even")
else:
    print("Odd")            # runs

# if-elif-else chain
marks = 82
if   marks >= 90: grade = "A+"
elif marks >= 80: grade = "A"   # ← this runs
elif marks >= 70: grade = "B"
elif marks >= 60: grade = "C"
else:             grade = "F"
print(grade)       # A
```

```python
#Nested if & ternary
#Nested if — check multiple conditions
x = 15
if x > 0:
    if x % 2 == 0:
        print("Positive even")
    else:
        print("Positive odd")   # runs
else:
    print("Non-positive")

# Ternary (one-liner if-else)
score = 55
result = "Pass" if score >= 40 else "Fail"
print(result)   # Pass

# Multiple conditions with and / or
age, has_id = 20, True
if age >= 18 and has_id:
    print("Entry allowed")   # runs

temp = 38
if temp < 10 or temp > 35:
    print("Extreme temperature!")  # runs
```

```python
#for loop — iterate over a sequence
# Over a list
fruits = ["apple", "mango", "banana"]
for fruit in fruits:
    print(fruit)
# apple  mango  banana

# Over a string
for ch in "Python":
    print(ch, end="-")
# P-y-t-h-o-n-

# range(start, stop, step)
for i in range(1, 10, 2):   # 1,3,5,7,9
    print(i, end=" ")

# Countdown with negative step
for i in range(5, 0, -1):   # 5,4,3,2,1
    print(i, end=" ")
```

```python
#Useful for-loop patterns
# enumerate — get index AND value
students = ["Alice", "Bob", "Carol"]
for i, name in enumerate(students, start=1):
    print(f"{i}. {name}")
# 1. Alice  2. Bob  3. Carol

# zip — loop two lists together
names  = ["Alice", "Bob"]
scores = [85, 92]
for n, s in zip(names, scores):
    print(f"{n}: {s}")

# dict.items() — loop key-value pairs
info = {"name":"Alice", "age":20}
for key, val in info.items():
    print(f"{key} = {val}")

# list comprehension — compact for loop
squares = [x**2 for x in range(1, 6)]
# [1, 4, 9, 16, 25]

# for-else — else runs if no break occurred
for i in range(5):
    print(i)
else:
    print("Loop finished normally")
```

```python
#while loop — repeat while condition is true
# Basic while
count = 1
while count <= 5:
    print(count)
    count += 1       # critical — must change condition!
# 1 2 3 4 5

# User input loop (sentinel pattern)
total = 0
while True:
    try:
        n_str = input("Enter number (0 to stop): ")
        if n_str.lower() == 'stop': # Allow 'stop' as a valid exit command
            n = 0
        else:
            n = int(n_str)
    except ValueError:
        print("Invalid input. Please enter a number or 'stop' to exit.")
        continue

    if n == 0:
        break
    total += n
print("Total:", total)

# while-else — else runs if no break
x = 10
while x > 0:
    x -= 3
else:
    print("x reached:", x)   # x = -2
```

```python
#Infinite loop with break
# Common pattern: while True + break
while True:
    cmd = input("Enter command (quit to exit): ")
    if cmd == "quit":
        print("Goodbye!")
        break
    print(f"You entered: {cmd}")

# Countdown timer
seconds = 5
while seconds > 0:
    print(f"{seconds}...")
    seconds -= 1
print("Blast off!")
```

```python
#break — exit loop immediately (loop control)
# Find first even number
nums = [3, 7, 9, 4, 11, 6]
for n in nums:
    if n % 2 == 0:
        print("First even:", n)   # First even: 4
        break   # stops here, 11 and 6 never checked

# Search in string
text = "Hello World"
for i, ch in enumerate(text):
    if ch == "W":
        print(f"Found W at index {i}")   # index 6
        break
```

```python
#continue — skip to next iteration (loop control)
# Skip negative numbers
data = [4, -2, 7, -5, 3]
for n in data:
    if n < 0:
        continue      # skip negatives
    print(n)          # 4  7  3

# Skip multiples of 3
for i in range(1, 11):
    if i % 3 == 0:
        continue
    print(i, end=" ")  # 1 2 4 5 7 8 10
```

```python
#Nested loops
# Multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end="\t")
    print()
# 1  2  3
# 2  4  6
# 3  6  9

# Star pattern (triangle)
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
#break in nested loops
# break only exits the INNERMOST loop
for i in range(3):          # outer: 0,1,2
    for j in range(5):      # inner: 0,1,2,3,4
        if j == 2:
            break           # exits inner only
        print(f"{i},{j}", end=" ")
    print()
# 0,0 0,1
# 1,0 1,1
# 2,0 2,1

# To break out of ALL loops — use a flag
found = False
for i in range(5):
    for j in range(5):
        if i * j > 6:
            found = True
            break
    if found: break
print(f"First pair: {i}, {j}")
```
