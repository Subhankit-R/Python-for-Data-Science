```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
num = 5
print(f"The factorial of {num} is {factorial(num)}")

num_negative = -3
print(f"The factorial of {num_negative} is {factorial(num_negative)}")

num_zero = 0
print(f"The factorial of {num_zero} is {factorial(num_zero)}")
```
